```python
import numpy as np
import pandas as pd

# ==== Факторы по районам (из Word-таблиц) ====
region_factors = {
    'Сино': {
        'Вакантные часы района': 540,
        'Доля учителей с высшим образованием': 0.88,
        'Доля учителей с неполным высшим': 0.07,
        'Доля женщин-учителей': 0.80,
        'Доля непедагогов': 0.01,
        'Количество учителей в районе': 5484
    },
    'Фирдавси': {
        'Вакантные часы района': 120,
        'Доля учителей с высшим образованием': 0.82,
        'Доля учителей с неполным высшим': 0.10,
        'Доля женщин-учителей': 0.77,
        'Доля непедагогов': 0.006,
        'Количество учителей в районе': 3550
    },
    'Шохмансур': {
        'Вакантные часы района': 263,
        'Доля учителей с высшим образованием': 0.83,
        'Доля учителей с неполным высшим': 0.08,
        'Доля женщин-учителей': 0.79,
        'Доля непедагогов': 0.007,
        'Количество учителей в районе': 2265
    },
    'Исмоили Сомони': {
        'Вакантные часы района': 0,
        'Доля учителей с высшим образованием': 0.85,
        'Доля учителей с неполным высшим': 0.12,
        'Доля женщин-учителей': 0.75,
        'Доля непедагогов': 0.004,
        'Количество учителей в районе': 2029
    }
}

# ==== Количество школ по районам (сумма = 163) ====
schools_per_region = {
    'Сино': 62,
    'Фирдавси': 41,
    'Шохмансур': 33,
    'Исмоили Сомони': 28
}

# ==== Генерация каркаса школ ====
rows = []
for region, n_schools in schools_per_region.items():
    vals = region_factors[region]
    
    # распределяем учителей и вакантные часы по школам
    teachers_dist = np.random.multinomial(vals['Количество учителей в районе'], [1/n_schools]*n_schools)
    vacant_dist = np.random.multinomial(vals['Вакантные часы района'], [1/n_schools]*n_schools)
    
    for sid in range(n_schools):
        rows.append({
            'Район': region,
            'ID школы': f"{region}_{sid+1}",
            'Учителя_в_школе': teachers_dist[sid],
            'Вакантные часы (школы)': vacant_dist[sid],
            # варьируем кадровые показатели вокруг районного среднего
            'Доля учителей с высшим образованием': np.clip(np.random.normal(vals['Доля учителей с высшим образованием'], 0.03), 0, 1),
            'Доля учителей с неполным высшим': np.clip(np.random.normal(vals['Доля учителей с неполным высшим'], 0.02), 0, 1),
            'Доля женщин-учителей': np.clip(np.random.normal(vals['Доля женщин-учителей'], 0.05), 0, 1),
            'Доля непедагогов': np.clip(np.random.normal(vals['Доля непедагогов'], 0.003), 0, 1),
            'Нагрузка на учителя': np.random.normal(18, 2)
        })

df_schools = pd.DataFrame(rows)

# ==== Доля вакантных часов на уровне школы ====
df_schools['Доля вакантных часов'] = df_schools['Вакантные часы (школы)'] / df_schools['Учителя_в_школе'].replace(0,1)

# ==== Средние значения предметов по районам ====
subject_means = {
    'Исмоили Сомони': {
        'Английский язык': 7.19, 'Русский язык': 7.15,
        'Математика': 6.85, 'Химия': 6.80, 'Биология': 6.75, 'Физика': 6.70
    },
    'Сино': {
        'Английский язык': 6.80, 'Русский язык': 6.75,
        'Математика': 7.01, 'Химия': 7.18, 'Биология': 7.05, 'Физика': 7.10
    },
    'Фирдавси': {
        'Английский язык': 6.85, 'Русский язык': 6.80,
        'Математика': 6.90, 'Химия': 7.10, 'Биология': 7.01, 'Физика': 6.95
    },
    'Шохмансур': {
        'Английский язык': 6.70, 'Русский язык': 6.65,
        'Математика': 6.80, 'Химия': 7.00, 'Биология': 6.95, 'Физика': 7.09
    }
}

# ==== Генерация предметов с округлением ====
for subj in ['Английский язык','Русский язык','Математика','Химия','Биология','Физика']:
    df_schools[subj] = df_schools.apply(
        lambda row: np.round(np.random.normal(subject_means[row['Район']][subj], 0.1), 3),
        axis=1
    )

# ==== Профиль школы ====
stem_mean = df_schools[['Математика','Физика','Химия']].mean(axis=1)
hum_mean = df_schools[['Английский язык','Русский язык']].mean(axis=1)
df_schools['Профиль школы'] = np.where(
    abs(stem_mean - hum_mean) < 0.5, 'Сбалансированный профиль',
    np.where(stem_mean > hum_mean, 'Технический профиль', 'Гуманитарный профиль')
)

# ==== Качество кадров ====
df_schools['Качество кадров'] = pd.cut(
    df_schools['Доля учителей с высшим образованием'],
    bins=[0, 0.7, 0.85, 1],
    labels=['Низкое','Среднее','Высокое']
)

# ==== Финальный порядок колонок ====
ordered_cols = [
    'ID школы','Район',
    'Английский язык','Русский язык','Математика','Химия','Биология','Физика',
    'Учителя_в_школе','Вакантные часы (школы)','Доля вакантных часов',
    'Доля учителей с высшим образованием','Доля учителей с неполным высшим',
    'Доля женщин-учителей','Доля непедагогов',
    'Нагрузка на учителя','Профиль школы','Качество кадров'
]

df_final = df_schools[ordered_cols]
```
